```python
import zipfile

zipfile.ZipFile('stanfordextra_v12.zip').extractall()

IMG_DIR = 'Images'  # Assuming images are in 'Images' folder; adjust if necessary

json_path = 'stanfordextra_v12/keypoints.json'  # Assuming the JSON is in the unzipped folder; adjust path if different

with open(json_path, 'r') as f:
    keypoints_data = json.load(f)

images = sorted(os.listdir(IMG_DIR))[:20]

mapped_keypoints = {}

for img_name in images:
    img_path = os.path.join(IMG_DIR, img_name)
    image = Image.open(img_path)
    # Assuming keypoints_data[img_name] is a list of [x, y] pairs
    kps = [Keypoint(x=k[0], y=k[1]) for k in keypoints_data.get(img_name, [])]
    kpsoi = KeypointsOnImage(kps, shape=image.size)
    mapped_keypoints[img_name] = kpsoi

# mapped_keypoints now contains the KeypointsOnImage for the first 20 images
```
